### **Convert weatehr3.csv into the following (mean value )**

```python
import pandas as pd
import matplotlib
import numpy as np
df=pd.read_csv(r"C:\Users\Dileesh\Desktop\python Exc\day4\1.csv")
df
```

### **PIVOT**

```python
df1 = df.pivot(index='city', columns='date', values=["temperature","humidity"])
df1
```

### **RESHAPE**

```python
df2=pd.read_csv(r"C:\Users\Dileesh\Desktop\python Exc\day4\2.csv")
df2
table = pd.pivot_table(df2, values=['temperature', 'humidity'],columns=['date'],index=['city'],aggfunc={'temperature': np.mean,'humidity': np.mean})
table


```

```python
table.to_csv("reshape.csv")
table.to_excel("reshape.xlsx")
```
